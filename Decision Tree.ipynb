### 1. Label Encoding (For Ordinal Categories or Tree-Based Models)
Converts each category into a unique integer.
Good for decision trees, where the model can handle the ordering.Use this if:

Categories have no order, and you’re using tree-based models (like DecisionTree, RandomForest).

You want to quickly convert string data into integers.


EX:from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Age'] = le.fit_transform(df['Age'])  # Youth=2, Middle=1, Senior=0 (example)

###  2. One-Hot Encoding (Best for Non-Tree Models like Logistic Regression, SVM)
Creates a new binary column for each category.
Use this if:

The model is not tree-based (e.g., linear models).

You want to avoid giving any ordinal meaning to categories.

The number of unique values per column is not too high.

EX:df_encoded = pd.get_dummies(df, columns=['Age', 'Income', 'Student', 'Credit_rating'], drop_first=True)

```python
Age_=LabelEncoder()
Income_=LabelEncoder()
Student_=LabelEncoder()
Credit_rating_=LabelEncoder()
Buys_computer_=LabelEncoder()
```

```python
df["Age_n"]=Age_.fit_transform(df["Age"])
df["Income_n"]=Income_.fit_transform(df["Income"])
df["Student_n"]=Student_.fit_transform(df["Student"])
df["Credit_rating_n"]=Credit_rating_.fit_transform(df["Credit_rating"])
df["Buys_computer_n"]=Buys_computer_.fit_transform(df["Buys_computer"])
```
