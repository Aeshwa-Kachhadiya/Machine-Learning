```python
import pandas as pd

# Sample data
data = {
    'ID': [1, 2, 3, 4],
    'Color': ['Red', 'Blue', 'Green', 'Red']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# One Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Color'])

print("\nAfter One Hot Encoding:")
print(df_encoded)
```

```python
import pandas as pd
import numpy as np
```

```python
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top_10=[x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
for label in top_10:
    df[label]=np.where(df["X2"]==label,1,0)

df[['X2']+top_10].head()
```

```python
def one_hot_top_x(data,variable,top_x_labels):
    
    for label in top_x_labels:
        data[variable+"_"+label]=np.where(df[variable]==label,1,0)

df=pd.read_excel("MercedesBenz.xlsx", usecols=["X1","X2","X3","X4","X5","X6"])

one_hot_top_x(df,"X2",top_10)
df.head()
```

```python
top_10=[x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]

df=pd.read_excel("MercedesBenz.xlsx", usecols=["X1","X2","X3","X4","X5","X6"])

one_hot_top_x(df,"X1",top_10)
df.head()
```
