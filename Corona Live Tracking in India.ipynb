```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_json("https://covid19dashboard.mohfw.gov.in/data/datanew.json")
df = df.drop(columns=['sno']).iloc[:-1]
```

```python
df.head()
```

```python
total_states=np.arange(len(df["state_name"]))
total_states
```

```python
#Total positive cases based on states in india
plt.figure(figsize=(9,6),dpi=80, facecolor="w", edgecolor="k")
plt.barh(total_states,df["positive"],align="center",alpha=0.5,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df["state_name"])
plt.xlim(1,max(df["positive"])+ 100)
plt.xlabel("Positive number of cases")
plt.ylabel("Corona virus cases")
plt.show()
```

```python
#Total active cases based on states in india
plt.figure(figsize=(9,6),dpi=80, facecolor="w", edgecolor="k")
plt.barh(total_states,df["active"],align="center",alpha=0.5,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df["state_name"])
plt.xlim(1,max(df["active"])+ 100)
plt.xlabel("active number of cases")
plt.ylabel("Corona virus cases")
plt.show()
```

```python
#Total death cases based on states in india
plt.figure(figsize=(9,6),dpi=80, facecolor="w", edgecolor="k")
plt.barh(total_states,df["death"],align="center",alpha=0.5,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df["state_name"])
plt.xlim(1,max(df["death"])+ 100)
plt.xlabel("death number of cases")
plt.ylabel("Corona virus cases")
plt.show()
```
